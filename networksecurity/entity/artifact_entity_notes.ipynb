# 📝 Notes — `@dataclass` in Python

---

## 🔹 What is `@dataclass`?
- `@dataclass` is a **decorator** in Python
- It automatically creates common methods like `__init__`, `__str__` for a class
- It saves you from writing **repetitive boilerplate code**

---

## 🆚 Without vs With `@dataclass`

### ❌ Without `@dataclass` (Long Way)
```python
class DataIngestionArtifact:
    def __init__(self, trained_file_path: str, test_file_path: str):
        self.trained_file_path = trained_file_path
        self.test_file_path = test_file_path
```

### ✅ With `@dataclass` (Short Way)
```python
@dataclass
class DataIngestionArtifact:
    trained_file_path: str
    test_file_path: str
```
> Both do the **exact same thing!**
> `@dataclass` just makes it **shorter and cleaner** ✅

---

## 🔹 What Does This Specific Code Do?

```python
from dataclasses import dataclass

@dataclass
class DataIngestionArtifact:
    trained_file_path: str   # Path to training data file
    test_file_path: str      # Path to testing data file
```

- This class is like a **container/box** that holds 2 file paths
- `trained_file_path` → path of the file used for **training the ML model**
- `test_file_path` → path of the file used for **testing the ML model**

---

## 🤔 Why is it Needed?

In ML projects, after **Data Ingestion** (loading data), we need to pass file paths to the next step. Instead of passing them separately, we wrap them in one object:

```python
# After data ingestion is done, create an artifact object
artifact = DataIngestionArtifact(
    trained_file_path="data/train.csv",
    test_file_path="data/test.csv"
)

# Now pass this single object to the next step
print(artifact.trained_file_path)  # data/train.csv
print(artifact.test_file_path)     # data/test.csv
```

---

## 💡 Simple Analogy
> Think of it like a **delivery package** 📦
> - Instead of carrying `train.csv` and `test.csv` separately
> - You pack both into **one box (artifact)**
> - And pass that one box to the next step in your ML pipeline

---

## ✅ Benefits of `@dataclass`

| Benefit | Meaning |
|---|---|
| Less code | No need to write `__init__` manually |
| Clean & readable | Easy to understand what data it holds |
| Auto `__str__` | Can print object directly |
| Reusable | Can be passed between functions easily |

---

## ⚡ One Line Summary
> *`@dataclass` is a shortcut to create simple data-holding classes without writing repetitive code* 🚀
